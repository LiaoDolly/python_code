```python
#印出連續大頁，美一頁每一個工作連結點進去後，列出工作技能，擅長工具與其他條件

import json
import time
from bs4 import BeautifulSoup
from collections import Counter
import requests

data = []


for i in range(1,5):
    #print('============================================================================'+'第'+str(i)+'頁')
    res=requests.get('https://www.104.com.tw/jobbank/joblist/joblist.cfm?jobsource=n104bank1\
    &ro=0&keyword=軟體程式設計師&order=2&asc=0&page='+str(i)+'&psl=N_A')

    
    
    soup = BeautifulSoup(res.text,'lxml').select('.job_name > a')
   
    for link in soup: 
        if 'href' in link.attrs:
            row1 = 'https://www.104.com.tw'+link.attrs['href']   

            res1 = requests.get(row1)
            soup1 = BeautifulSoup(res1.text,'lxml')
            
            jobInfo = dict()             #dictionry 每次迴圈進來都要淨空，不然放進List的東西會很雜
            jobInfo["title"] = soup1.select('h1')[1].contents[0].replace("\r\n","").replace(" ","")
            jobInfo["company"] = soup1.select('.cn')[0].text
            jobInfo["content"] = soup1.select_one("div.content > p").text.replace("\r", "").replace(" ", "")
            jobInfo["tools"] = soup1.select('.tool')[0].text
            jobInfo["skills"] = soup1.select('.tool')[1].text
            jobInfo["post_date"]= soup1.select('.update')[0].text
            jobInfo["other_condition"] = soup1.select('dl')[1].select('dd')[-1].text.replace("\r", "").replace(" ", "")
            
            try:
                data.append(jobInfo)
                time.sleep(5)
            
                print(len(data))   #印記入的次數就好 
            
            except Exception as e:
                print(e)
                pass
            
            
            
#print(data)
with open('Carters104_軟體程式設計師-1-4.json','w') as f:
    json.dump(data,f)  
 
         

#             print('====================================')
            
                
      
```
