```python
# crawling all detailed job info in one company

import requests
from bs4 import BeautifulSoup

for i in range(1,10):
    url = "https://www.104.com.tw/jobbank/custjob/index.php?r=cust&jobsource=checkc&j=623a436e5a46402140323c1d1d1d1d5f2443a363189j50&page="+str(i)+"#info06"     
    res = requests.get(url)
    print('============================================================================'+'第'+str(i)+'頁')
    # res.encoding       #check which code does this website use
    # res.status_code    # check ssh between user and webserver
    res.encoding = 'utf-8'   # the way to set code we want to use
    soup = BeautifulSoup(res.text,'lxml').select('.jobname > a')
    
    for link in soup: 
        if 'href' in link.attrs:
            row1 = 'https://www.104.com.tw'+link.attrs['href']   
            #print(type(row1))  #String type
            print(row1)        #many links
            
            
            
            res1 = requests.get(row1)
            soup2 = BeautifulSoup(res1.text,'lxml').select('.tool')
            soup3 = BeautifulSoup(res1.text,'lxml').select('dl')[1].select('dd')[-1]
            soup4 = BeautifulSoup(res1.text,'lxml').select('h1')
            soupx = soup4[1].contents
            soup5 = BeautifulSoup(res1.text,'lxml').select('.update')
            
            print(soupx[0].replace("\r\n","").replace(" ",""))   #jobname
            n = len(soup2)                                       #擅長工具
            for q in range(n):
                print(soup2[q].get_text())    
                print('----')
                

            soup4 = soup3.text.replace('\r','\n')                #其他條件
                    
            print(soup4)
            print(soup5[0].get_text())
            print("==============================================")
        
   
```

# (4)crawling page by page. focusing on the competency mentioned  frequently

```python
#印出連續大頁，美一頁每一個工作連結點進去後，列出工作技能，擅長工具與其他條件
#
#
#
import time
from bs4 import BeautifulSoup
from collections import Counter
import requests

need = Counter()

for i in range(1,5):
    print('============================================================================'+'第'+str(i)+'頁')
    res=requests.get('https://www.104.com.tw/jobbank/joblist/joblist.cfm?keyword=%E7%A8%8B%E5%BC%8F%E8%A8%AD%E8%A8%88%E5%B8%AB&jobsource=n104bank1&ro=0&order='+str(i))

    
    soup1 = BeautifulSoup(res.text,'lxml').select('.job_name > a')
    #print(soup)
    for link in soup1: 
        if 'href' in link.attrs:
            row1 = 'https://www.104.com.tw'+link.attrs['href']   
            #print(type(row1))            #String type
            print(row1)                   #many links
        
        
            
            res1 = requests.get(row1)        
            
            soup2 = BeautifulSoup(res1.text,'lxml').select('.tool')
            soup3 = BeautifulSoup(res1.text,'lxml').select('dl')[1].select('dd')[-1]
            soup5 = BeautifulSoup(res1.text,'lxml').select('.cn')
            soup6 = BeautifulSoup(res1.text,'lxml').select('h1')
            soup7 = BeautifulSoup(res1.text,'lxml').select('.update')
            
            print(soup5[0].get_text())                             #company name
            
            soupx = soup6[1].contents                             # job name
            print(soupx[0].replace("\r\n","").replace(" ",""))
        
            n = len(soup2)                                      #擅長工具
            for q in range(n):
                print(soup2[q].get_text())    
                print('----')
                

            soup4 = soup3.text.replace('\r','\n')                 #其他條件
            time.sleep(20)      
            #print(soup4)
            print(soup7[0].get_text())                          #更新時間
            print('====================================')
            
                
      
```

```python
#印出連續大頁，每一個工作連結點進去後，要個工作技能與其他條件，統計它！！
import time
from bs4 import BeautifulSoup
from collections import Counter
import requests

cnt = Counter()

for i in range(1,151):
   
    print(i)
    res=requests.get('https://www.104.com.tw/jobbank/joblist/joblist.cfm?keyword=%E7%A8%8B%E5%BC%8F%E8%A8%AD%E8%A8%88%E5%B8%AB&jobsource=n104bank1&ro=0&order='+str(i))

    
    soup1 = BeautifulSoup(res.text,'lxml').select('.job_name > a')
  
    for link in soup1: 
        #time.sleep(5)
        if 'href' in link.attrs:
            row1 = 'https://www.104.com.tw'+link.attrs['href']   
            
            
            res1 = requests.get(row1)
            soup2 = BeautifulSoup(res1.text,'lxml').select('.tool')
            
            #soup3 = BeautifulSoup(res1.text,'lxml').select('dl')[1].select('dd')[-1]
            
            
            n = len(soup2)               #擅長工具
            for q in range(n):
                result = soup2[q].get_text().split('、')   #開始要截字拉
                #print(result)  
                for accu in result:
                    if accu in cnt:
                        cnt[accu] += 1
                    else:
                        cnt[accu] =1
        
                
print(cnt.most_common(20))
            #soup4 = soup3.text.replace('\r','\n')
                    
            #print(soup4)
```
