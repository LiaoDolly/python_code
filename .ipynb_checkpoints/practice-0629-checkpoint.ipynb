```python
with open('../Downloads/cai_speech.txt') as f:
    speech = f.read()
```

```python
wd_count = Counter()

for word in jieba.cut(speech):
    print(word)
    if word in wd_count:
        wd_count[word] += 1
    else:
        wd_count[word] = 1
    
```

```python
wd_count.most_common(20)
```

```python
def sum(*args):
    res = 0
    for x in args:
        res =res + x
    return res
```

```python
sum(2,3,4,5,6,7,33)
```
