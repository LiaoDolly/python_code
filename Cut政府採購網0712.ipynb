```python
data_str1="""method:search
searchMethod:true
searchTarget:TPAM
orgName:
orgId:
hid_1:1
tenderName:
tenderId:
tenderType:tenderDeclaration
tenderWay:
tenderDateRadio:on
tenderStartDateStr:106/07/12
tenderEndDateStr:106/07/12
tenderStartDate:106/07/12
tenderEndDate:106/07/12
isSpdt:N
spdtStartDate:
spdtEndDate:
tenderSpdtYmStart:
tenderSpdtYmEnd:
opdtStartDate:
opdtEndDate:
proctrgCate:
tenderRange:
minBudget:
maxBudget:
location:
locationPredict:
priorityCate:
isReConstruct:
btnQuery:查詢"""
```

```python
data_str1
```

```python
def str2dict(data_str):
    data = {}
    for row in data_str.split('\n'):
        kv_list = row.split(":")
        data[kv_list[0]] = kv_list[1]
    return data
```

```python
data1 = str2dict(data_str1)
```

```python
data1
```

```python
res = requests.post('http://web.pcc.gov.tw/tps/pss/tender.do?searchMode=common&searchType=advance',data=data1)
```
